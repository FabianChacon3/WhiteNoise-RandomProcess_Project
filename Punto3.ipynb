Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter, correlate, welch
```

Punto 3

```python

x = np.random.normal(0, 1, 10000) #Señal aleatoria de media=0 , varianza=1 y 10.000 muestras
a = 2  # mediana
b = [1, -1/a, 1/2*a]
y = lfilter(b, [1], x)
mean_y = np.mean(y)
var_y = np.var(y)
autocorr_y = correlate(y, y, mode='full') / len(y)
plt.plot(autocorr_y[len(y)-100:len(y)+100])
plt.title("Autocorrelación")
plt.show()
```
